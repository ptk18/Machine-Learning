```python
import numpy as np
```

```python
class Perceptron():
  def __init__(self, weight_init=[-1,0.3,-0.2], bias=1, learning_rate=0.2, epoch=4, verbose=False):
    self.verbose = verbose
    self.W = np.array(weight_init)
    self.b = bias
    self.lr = learning_rate
    self.epoch = epoch

  def aggregation_fn(self, x):
    x = np.insert(x, 0, self.b)
    z = self.W.T.dot(x)
    return z

  def activation_fn(self,y):
    return 1 if y >= 0 else 0

  def predict(self,x):
    z = self.aggregation_fn(x)
    z = round(z,2)
    y = self.activation_fn(z)
    return y

  def fit(self, X, y_desired):
    for step in range(self.epoch):
      if self.verbose:
        print("Epoch=", step)
      for i in range(len(y_desired)):
        x = X[i]
        y = self.predict(x)
        e = y_desired[i] - y
        x = np.insert(x, 0, self.b)
        self.W = self.W + self.lr*e*x
        if self.verbose:
          print("the new weight (w0, w1, ...) =",self.W)
```

```python
import pandas as pd

filename = 'AND_gate.csv'
#filename = 'OR_gate.csv'
#filename = 'XOR_gate.csv'

df = pd.read_csv(filename)

X = df.drop('y', axis=1).values
y=df['y'].values
```

```python
model = Perceptron(learning_rate=0.2, epoch=16, verbose=True)
model.fit(X,y)
```

```python
misclassified = 0
for i in range(len(y)):
  x_test = X[i]
  y_model = model.predict(x_test)
  if y[i] != y_model:
    misclassified+=1
  print("actual=", y[i], " ,model=", y_model)
print("misclassified=", misclassified, ", out of ", len(y))
```
